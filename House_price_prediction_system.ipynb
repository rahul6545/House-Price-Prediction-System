```python
cat_fill_none=["Mas Vnr Type", "Fireplace Qu", "Garage Type", "Garage Finish", "Garage Qual", "Garage Cond", "Bsmt Qual", "Bsmt Cond", "Bsmt Exposure", "BsmtFin Type 1", "BsmtFin Type 2"]

for col in cat_fill_none:

        if col in df.columns:

            df.fillna({col: 'None'},inplace=True)
```

```python
num_fill_median = ["Lot Frontage","Garage Yr Blt","Mas Vnr Area"]
for col in num_fill_median:
    if col in df.columns:
        df.fillna({col:df[col].median()},inplace=True)
```
